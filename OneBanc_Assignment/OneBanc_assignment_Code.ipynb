```python
import pandas as pd
import numpy as np
```

```python
#reading csv file (it has missing values)
col_names=['Date', 'Description', 'International', 'Category', 'Amount', 'User', 'Transaction Type', 'Currency', 'Locality', 'City', 'State', 'Country']
data_raw=pd.read_csv('/content/CleanStatements.csv', names=col_names)
data_raw
```

```python
#dropping top row header
data_raw=data_raw.loc[1:]
```

```python
#dropping the rows where NAN value is present
data_all=data_raw.dropna()
```

```python
data_all=data_all.reset_index(drop=True)
data_all
```

```python
#extracting amount column values
amount_list=list(data_all['Amount'])
print(amount_list)
```

```python
#stripping each value (removing white space)
amount_list2=[]
for val in amount_list:
  amount_list2.append(val.strip())

print(amount_list2)
```

```python
#storing the index of row where amount value is not a number
index_to_drop=[]

for i in range(len(amount_list2)):

  hold= amount_list2[i]

  if hold[0].isalpha():
    index_to_drop.append(i)
  


print(index_to_drop)
```

```python
data_all['Amount']=amount_list2
data_all
```

```python
#dropping the rows with non numeric amount values
data_all=data_all.drop(index_to_drop, axis=0)
data_all=data_all.reset_index(drop=True)
data_all
```

```python
#conveting the amount into positive numeric form
data_all['Amount']=pd.to_numeric(data_all['Amount'])
data_all['Amount']=data_all['Amount'].abs()
data_all
```

```python
#converting the date into date-time format from string type
data_all['Date']=pd.to_datetime(data_all['Date'])
data_all.head(20)
```

```python
#extracting categories from the data
list_cat = list(data_all['Category'])
print(list_cat)
```

```python
#stripping white-space from the category
list_cat2=[]

for cat in list_cat:
  list_cat2.append(cat.strip())

print(list_cat2)
```

```python
#finding already existing uniue categories

dict_category_intial={}

for cat in list_cat2:

  if cat in dict_category_intial:
    dict_category_intial[cat]+=1
  else:
    dict_category_intial[cat]=1


print(len(dict_category_intial))
```

```python
#print categories with their counts
dict_category_intial
```

```python
#dictionary of words used in category column
dict_cat_words={}

for cat in dict_category_intial:
  #list of words (space separated)
  hold= cat.split(" ")
  
  for word in hold:

    if word in dict_cat_words:
      dict_cat_words[word]+=1
    else:
      dict_cat_words[word]=1
```

```python
print(len(dict_cat_words))
```

```python
#defining list of keywords for new categories
#adding some common words as keywords (generic keywords)

entertainment=['game','film', 'song', 'cinema', 'song', 'entertainment','television','actor', 'actress']
food = ['food', 'restaurant', 'cafe', 'dish', 'pizza', 'burger', 'chips', 'drinks', 'mocktail']
travel = ['city', 'hotel', 'lake', 'highway', 'railway', 'visit', 'palace', 'sea', 'beach', 'cruise', 'travel']
cab = ['ridesharing', 'ride', 'ola', 'uber', 'cab']
shopping = ['e-commerce','retail', 'fashion', 'clothing', 'apparel', 'flipkart', 'amazon', 'buy', 'price', 'shop', 'gifts', 'gift', 'purchase', 'supermarket', 'mobile', 'phone', 'electronics','store']
book = ['book', 'novel', 'journal', 'autobiography', 'poetry', 'poem', 'ebook']
flight = ['flight','economy', 'business', 'lufthansa', 'air', 'plane', 'hostess', 'crew', 'pilot']
```

```python
#categorizing each previous category into new using keywords
#default catgeory is miscellaneous

list_cat3=[]

#converting in lowercase (for comparison)
for cat in list_cat2:

  list_cat3.append(cat.lower())

print(list_cat3)
```

```python
print((list_cat3[0].strip(" ")))
```

```python
#forming new list of category

list_category_new=[]

for cat in list_cat3:

  hold = cat.split(" ")
  
  flag=0
  for word in hold:

    if word in entertainment:

      list_category_new.append("entertainment")
      flag=1
      break

    elif word in food:

      list_category_new.append("food")
      flag=1
      break

    elif word in travel:

      list_category_new.append("travel")
      flag=1
      break

    elif word in flight:

      list_category_new.append("flight")
      flag=1
      break

    elif word in cab:

      list_category_new.append("cab")
      flag=1
      break

    elif word in shopping:

      list_category_new.append("shopping")
      flag=1
      break

    elif word in book:

      list_category_new.append("book")
      flag=1
      break

  

  if flag == 0:
    list_category_new.append("general")

print(list_category_new)
```

```python
print(len(list_category_new))
```

```python
#modifying category
data_all['Category']=list_category_new
```
